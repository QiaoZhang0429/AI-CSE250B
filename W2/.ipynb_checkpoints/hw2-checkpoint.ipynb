```python
import numpy as np
import math
import cPickle
import collections
```

```python
# load vocabulary
#vocabulary = collections.defaultdict(int)
vocabulary = []
fv = open('vocabulary.txt', 'r')
for line in fv.readlines():
    b = line.strip()
    vocabulary.append(b)
    #if vocabulary.has_key(b):
    #    print "error"
    #    continue
    #vocabulary[b] = len(vocabulary)
V = len(vocabulary)
```

```python
print V
print vocabulary
```

```python
# load label
train_label = []
fl = open('20news-bydate-matlab/matlab/train.label', 'r')
for line in fl.readlines():
    b = int(line.strip())
    train_label.append(b)
docNumber = len(train_label)
```

```python
print docNumber
print train_label
```

```python
# load matrix
train_matrix = np.zeros([docNumber, V])
fd = open('20news-bydate-matlab/matlab/train.data', 'r')
for line in fd.readlines():
    b = line.strip().split()
    if(len(b) != 3):
        print "error"
    train_matrix[int(b[0])-1][int(b[1])-1] = int(b[2])
```

```python
print train_matrix.shape
print train_matrix[0][50000:50050]
```

```python
# calculate \pi_j
pi = collections.defaultdict(int)
for i in range(len(train_label)):
    if pi.has_key(train_label[i]):
        pi[train_label[i]] += 1
    else:
        pi[train_label[i]] = 1
for key in pi:
    pi[key] = pi[key] * 1.0 / docNumber
```

```python
print pi
```

```python
# n_j
nj = collections.defaultdict(int)
for i in range(len(train_label)):
    if nj.has_key(train_label[i]):
        nj[train_label[i]] += np.sum(train_matrix[i])
    else:
        nj[train_label[i]] = np.sum(train_matrix[i])
```

```python
print nj
```

```python
# n_{jw}
K = 20
n_jw = np.zeros([K, V])
for i in range(docNumber):
    for j in range(V):
        if (i * j) % 100000 == 0:
            print i * j * 1.0 / (docNumber * V)
        n_jw[train_label[i]-1][j] += train_matrix[i][j]
```

```python
print n_jw[0][:50]
```

```python
# calculate p_{jw} with smoothing
prob = np.zeros([K, V])
for i in range(K):
    for j in range(V):
        prob[i][j] = (n_jw[i][j]+1) * 1.0 / (nj[i+1] + V)
```

```python
print prob[0][:50]
```
